```python
df["timestamp"] = pd.to_datetime(
    df["year"].astype(str) + df["day_of_year"].astype(str).str.zfill(3),
    format="%Y%j"
) + pd.to_timedelta(df["hour"], unit="h")

df[["timestamp", "year", "day_of_year", "hour"]].head()
```

The interplanetary magnetic field, or IMF, is one of the key drivers of space weather because it controls how solar wind energy interacts with Earth’s magnetic environment. The orientation of the IMF, especially the north–south Bz component, plays a major role in determining whether solar activity produces mild auroras or stronger geomagnetic storms. When Bz turns southward/negative, it can connect more effectively with Earth’s magnetic field, allowing more solar wind energy to enter the magnetosphere. This increases the likelihood of geomagnetic disturbance and auroral activity.

```python
# NASA OMNIWeb uses special numbers for missing values
missing_values = {
    "imf_mag_nt": 999.9,
    "by_gsm_nt": 999.9,
    "bz_gsm_nt": 999.9,
    "proton_temp_k": 9999999,
    "proton_density_ncc": 999.9,
    "flow_speed_kms": 9999,
    "flow_pressure_npa": 99.99,
    "electric_field_mvm": 999.99,
    "ae_nt": 9999,
    "dst_nt": 99999
}

for col, missing_val in missing_values.items():
    if col in df.columns:
        df[col] = df[col].replace(missing_val, np.nan)

# Check missing values
df.isnull().sum()
```

## 5. Create Clean Modeling Dataset

Rows with missing values are removed for the first version of the project.  
Later, missing-value imputation can be added if needed.

```python
df_clean = df.dropna().copy()

print("Original shape:", df.shape)
print("Clean shape:", df_clean.shape)

df_clean.head()
```

## 6. Create Aurora Intensity Labels

The Kp index is converted into three interpretable intensity groups:

- Quiet: Kp < 3
- Moderate: 3 <= Kp < 5
- Storm / Aurora Likely: Kp >= 5

```python
def classify_kp(kp):
    if kp < 3:
        return "Quiet"
    elif kp < 5:
        return "Moderate"
    else:
        return "Storm / Aurora Likely"

df_clean["intensity_label"] = df_clean["kp"].apply(classify_kp)

df_clean[["timestamp", "kp", "aurora_visible", "intensity_label"]].head()
```

## 7. Check Aurora Class Balance

This step checks how many observations belong to aurora-likely and non-aurora conditions.

```python
class_counts = df_clean["aurora_visible"].value_counts()
class_percentages = df_clean["aurora_visible"].value_counts(normalize=True) * 100

print("Class counts:")
print(class_counts)

print("\nClass percentages:")
print(class_percentages)
```

## 8. Create Time-Series Lag and Rolling Features

Aurora activity depends not only on current space-weather conditions, but also on previous conditions.  
Lag and rolling-window features help the model learn short-term space-weather behavior.

```python
df_clean = df_clean.sort_values("timestamp").copy()

# Lag features
df_clean["bz_lag_1h"] = df_clean["bz_gsm_nt"].shift(1)
df_clean["bz_lag_3h"] = df_clean["bz_gsm_nt"].shift(3)
df_clean["bz_lag_6h"] = df_clean["bz_gsm_nt"].shift(6)

df_clean["speed_lag_1h"] = df_clean["flow_speed_kms"].shift(1)
df_clean["speed_lag_3h"] = df_clean["flow_speed_kms"].shift(3)
df_clean["speed_lag_6h"] = df_clean["flow_speed_kms"].shift(6)

df_clean["density_lag_1h"] = df_clean["proton_density_ncc"].shift(1)
df_clean["density_lag_3h"] = df_clean["proton_density_ncc"].shift(3)

# Rolling features
df_clean["bz_rolling_3h"] = df_clean["bz_gsm_nt"].rolling(3).mean()
df_clean["bz_rolling_6h"] = df_clean["bz_gsm_nt"].rolling(6).mean()

df_clean["speed_rolling_3h"] = df_clean["flow_speed_kms"].rolling(3).mean()
df_clean["speed_rolling_6h"] = df_clean["flow_speed_kms"].rolling(6).mean()

df_clean["density_rolling_3h"] = df_clean["proton_density_ncc"].rolling(3).mean()

df_model = df_clean.dropna().copy()

print("Modeling dataset shape:", df_model.shape)
df_model.head()
```

## 9. Define Leakage-Aware Features Without AE

Current-hour `ae_nt` is excluded because the Auroral Electrojet Index directly measures auroral-zone activity and may create leakage-like behavior.  
`kp` and `kp_x10` are also excluded because the target variable `aurora_visible` was created from Kp.

```python
features = [
    "imf_mag_nt",
    "by_gsm_nt",
    "bz_gsm_nt",
    "proton_temp_k",
    "proton_density_ncc",
    "flow_speed_kms",
    "flow_pressure_npa",
    "electric_field_mvm",
    "dst_nt",
    "bz_lag_1h",
    "bz_lag_3h",
    "bz_lag_6h",
    "speed_lag_1h",
    "speed_lag_3h",
    "speed_lag_6h",
    "density_lag_1h",
    "density_lag_3h",
    "bz_rolling_3h",
    "bz_rolling_6h",
    "speed_rolling_3h",
    "speed_rolling_6h",
    "density_rolling_3h"
]

X = df_model[features]
y = df_model["aurora_visible"]

print("Feature matrix shape:", X.shape)
print("Target shape:", y.shape)
print("Features used:")
print(features)
```

## 10. Time-Aware Train-Test Split Without AE

Because this is time-series data, the model should not randomly mix past and future observations.

In this step, the first 80% of the dataset is used for training, and the final 20% is used for testing.  
This better reflects a real forecasting situation where the model learns from earlier space-weather patterns and is tested on later observations.

The feature set excludes:

- `ae_nt`, because it may create leakage-like behavior
- `kp` and `kp_x10`, because the target was created from Kp
- `aurora_visible`, because this is the prediction target

```python
# ------------------------------------------------------------
# Time-aware train-test split
# ------------------------------------------------------------
# Since the dataset is ordered by timestamp, we split by row order.
# First 80% of the observations are used for training.
# Last 20% of the observations are used for testing.

split_index = int(len(df_model) * 0.8)

# Training features and target
X_train = X.iloc[:split_index]
y_train = y.iloc[:split_index]

# Testing features and target
X_test = X.iloc[split_index:]
y_test = y.iloc[split_index:]

# Print dataset shapes
print("Train feature shape:", X_train.shape)
print("Test feature shape:", X_test.shape)
print("Train target shape:", y_train.shape)
print("Test target shape:", y_test.shape)

# Check class balance in train and test sets
print("\nTrain target distribution:")
print(y_train.value_counts())

print("\nTest target distribution:")
print(y_test.value_counts())
```

## 11. Train Baseline Random Forest Model Without AE

This section trains the first supervised machine learning model using Random Forest.

Random Forest is a strong baseline model because it can capture nonlinear relationships between space-weather variables and aurora-likely conditions.

The model uses `class_weight="balanced"` because aurora-likely events are less frequent than non-aurora conditions. This helps the model pay more attention to the minority class.

```python
# ------------------------------------------------------------
# Train baseline Random Forest model
# ------------------------------------------------------------

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Create Random Forest classifier
# n_estimators=300 means the model builds 300 decision trees.
# class_weight="balanced" helps with class imbalance.
rf_model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced"
)

# Train the model using the training data
rf_model.fit(X_train, y_train)

# Predict class labels on the test data
y_pred = rf_model.predict(X_test)

# Predict probability of aurora_visible = 1
# This is useful for ROC-AUC and threshold tuning later.
y_prob = rf_model.predict_proba(X_test)[:, 1]

# Print evaluation results
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC:")
print(roc_auc_score(y_test, y_prob))
```

## 12. Save Baseline Model Metrics

This section stores the baseline Random Forest results in a clean table.

The most important metric for this project is recall for class `1`, because class `1` represents aurora-likely conditions.  
A low recall means the model is missing aurora events.

```python
# ------------------------------------------------------------
# Save baseline model metrics in a dataframe
# ------------------------------------------------------------

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Create a clean results table for the baseline model
baseline_results = pd.DataFrame({
    "model": ["Random Forest Without AE"],
    "threshold": [0.50],
    "accuracy": [accuracy_score(y_test, y_pred)],
    "precision_aurora": [precision_score(y_test, y_pred, zero_division=0)],
    "recall_aurora": [recall_score(y_test, y_pred, zero_division=0)],
    "f1_aurora": [f1_score(y_test, y_pred, zero_division=0)],
    "roc_auc": [roc_auc_score(y_test, y_prob)]
})

# Display baseline results
baseline_results
```

```python
# =============================================================================
# AuroraAI — Advanced Threshold Tuning & NASA-Style Visualization
# Run these cells in Jupyter after Cell 12 (baseline_results).
# Requires: y_test, y_prob, rf_model, features, df_model, baseline_results
# =============================================================================

# ── Cell 13: Setup & Style Configuration ─────────────────────────────────────

import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
from matplotlib.ticker import MultipleLocator
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, precision_recall_curve, auc,
    confusion_matrix, classification_report
)

OUTPUT_DIR = r"C:\Users\User\Desktop\Aurora project\outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# NASA dark aurora palette — all hardcoded for matplotlib (no CSS vars)
VOID    = "#020508"
SPACE   = "#06111f"
NEBULA  = "#0a1e38"
GREEN   = "#00ff87"
TEAL    = "#00e5d4"
BLUE    = "#3b82f6"
VIOLET  = "#8b5cf6"
PINK    = "#f472b6"
ORANGE  = "#f97316"
YELLOW  = "#facc15"
WHITE   = "#e2e8f0"
DIM     = "#64748b"
MID     = "#94a3b8"
GRID    = "#0f2035"

def nasa_axes(ax, title="", xlabel="", ylabel=""):
    ax.set_facecolor(SPACE)
    ax.tick_params(colors=MID, labelsize=9)
    for sp in ax.spines.values():
        sp.set_color(NEBULA)
    ax.grid(color=GRID, linewidth=0.6, linestyle="--", alpha=0.7)
    ax.set_title(title, color=WHITE, fontsize=11, fontweight="bold",
                 pad=10, fontfamily="monospace")
    ax.set_xlabel(xlabel, color=MID, fontsize=9, fontfamily="monospace")
    ax.set_ylabel(ylabel, color=MID, fontsize=9, fontfamily="monospace")

def save_fig(fig, name):
    path = os.path.join(OUTPUT_DIR, name)
    fig.savefig(path, dpi=150, bbox_inches="tight", facecolor=VOID)
    plt.show()
    print(f"  Saved → {path}")

print("Cell 13 complete: output directory and style config ready.")
print(f"  Output path: {OUTPUT_DIR}")


# ── Cell 14: Threshold Sweep → threshold_df ───────────────────────────────────

thresholds = np.arange(0.05, 0.81, 0.05)
rows = []

for t in thresholds:
    y_pred_t = (y_prob >= t).astype(int)
    cm = confusion_matrix(y_test, y_pred_t)
    tn, fp, fn, tp = cm.ravel() if cm.shape == (2, 2) else (0, 0, 0, 0)
    rows.append({
        "threshold":        round(float(t), 2),
        "accuracy":         round(accuracy_score(y_test, y_pred_t), 6),
        "precision_aurora": round(precision_score(y_test, y_pred_t, zero_division=0), 6),
        "recall_aurora":    round(recall_score(y_test, y_pred_t, zero_division=0), 6),
        "f1_aurora":        round(f1_score(y_test, y_pred_t, zero_division=0), 6),
        "roc_auc":          round(roc_auc_score(y_test, y_prob), 6),
        "TP": int(tp), "TN": int(tn), "FP": int(fp), "FN": int(fn),
    })

threshold_df = pd.DataFrame(rows)

best_f1_idx   = threshold_df["f1_aurora"].idxmax()
best_threshold = threshold_df.loc[best_f1_idx, "threshold"]

print("=" * 58)
print("  THRESHOLD SWEEP RESULTS")
print("=" * 58)
print(threshold_df[["threshold","precision_aurora","recall_aurora",
                     "f1_aurora","accuracy"]].to_string(index=False))
print()
print(f"  Best F1 threshold : {best_threshold}")
print(f"  Best F1 score     : {threshold_df.loc[best_f1_idx, 'f1_aurora']}")
print(f"  Recall at best F1 : {threshold_df.loc[best_f1_idx, 'recall_aurora']}")


# ── Cell 15: Apply best_threshold → confusion matrix + report ────────────────

y_pred_optimal = (y_prob >= best_threshold).astype(int)

cm_opt = confusion_matrix(y_test, y_pred_optimal)
tn_o, fp_o, fn_o, tp_o = cm_opt.ravel()

print("=" * 58)
print(f"  OPTIMAL THRESHOLD: {best_threshold}")
print("=" * 58)
print("\nConfusion Matrix:")
print(f"  TN={tn_o}  FP={fp_o}")
print(f"  FN={fn_o}  TP={tp_o}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_optimal,
                            target_names=["No Aurora", "Aurora"]))

optimal_result = pd.DataFrame([{
    "model":            "Random Forest Without AE (Tuned)",
    "threshold":         best_threshold,
    "accuracy":          round(accuracy_score(y_test, y_pred_optimal), 6),
    "precision_aurora":  round(precision_score(y_test, y_pred_optimal, zero_division=0), 6),
    "recall_aurora":     round(recall_score(y_test, y_pred_optimal, zero_division=0), 6),
    "f1_aurora":         round(f1_score(y_test, y_pred_optimal, zero_division=0), 6),
    "roc_auc":           round(roc_auc_score(y_test, y_prob), 6),
}])

all_results = pd.concat([baseline_results, optimal_result], ignore_index=True)
print("\nBefore / After comparison:")
print(all_results[["model","threshold","precision_aurora",
                   "recall_aurora","f1_aurora","roc_auc"]].to_string(index=False))


# ── Cell 16: Chart 1 — ROC Curve ─────────────────────────────────────────────

fpr_vals, tpr_vals, roc_thresholds = roc_curve(y_test, y_prob)
roc_auc_val = auc(fpr_vals, tpr_vals)

fig, ax = plt.subplots(figsize=(7, 5.5), facecolor=VOID)
nasa_axes(ax, title="ROC Curve — Aurora Classification",
          xlabel="False Positive Rate", ylabel="True Positive Rate")

ax.fill_between(fpr_vals, tpr_vals, alpha=0.12, color=GREEN)
ax.plot(fpr_vals, tpr_vals, color=GREEN, lw=2.5,
        label=f"Random Forest (AUC = {roc_auc_val:.4f})")
ax.plot([0, 1], [0, 1], "--", color=DIM, lw=1, label="Random classifier")

# Mark current optimal threshold point on ROC
opt_idx = np.argmin(np.abs(roc_thresholds - best_threshold))
ax.scatter(fpr_vals[opt_idx], tpr_vals[opt_idx],
           color=ORANGE, s=90, zorder=5,
           label=f"Optimal threshold ({best_threshold})")
ax.annotate(f"  t={best_threshold}\n  TPR={tpr_vals[opt_idx]:.3f}\n  FPR={fpr_vals[opt_idx]:.3f}",
            xy=(fpr_vals[opt_idx], tpr_vals[opt_idx]),
            xytext=(fpr_vals[opt_idx]+0.08, tpr_vals[opt_idx]-0.12),
            color=ORANGE, fontsize=8, fontfamily="monospace",
            arrowprops=dict(arrowstyle="->", color=ORANGE, lw=1))

leg = ax.legend(facecolor=SPACE, labelcolor=WHITE, fontsize=9,
                edgecolor=NEBULA, loc="lower right")
ax.set_xlim(-0.02, 1.02)
ax.set_ylim(-0.02, 1.08)
fig.tight_layout(pad=1.5)
save_fig(fig, "01_roc_curve.png")


# ── Cell 17: Chart 2 — Precision-Recall Curve ────────────────────────────────

prec_vals, rec_vals, pr_thresholds = precision_recall_curve(y_test, y_prob)
pr_auc_val = auc(rec_vals, prec_vals)
no_skill   = y_test.mean()

fig, ax = plt.subplots(figsize=(7, 5.5), facecolor=VOID)
nasa_axes(ax, title="Precision-Recall Curve — Aurora Class",
          xlabel="Recall", ylabel="Precision")

ax.fill_between(rec_vals, prec_vals, alpha=0.12, color=TEAL)
ax.plot(rec_vals, prec_vals, color=TEAL, lw=2.5,
        label=f"Random Forest (PR-AUC = {pr_auc_val:.4f})")
ax.axhline(no_skill, color=DIM, lw=1, linestyle="--",
           label=f"No-skill baseline ({no_skill:.3f})")
ax.axvline(0.75, color=ORANGE, lw=1, linestyle=":",
           label="Recall = 0.75 target")

# Mark optimal threshold
opt_pr_idx = np.argmin(np.abs(pr_thresholds - best_threshold))
ax.scatter(rec_vals[opt_pr_idx], prec_vals[opt_pr_idx],
           color=ORANGE, s=90, zorder=5,
           label=f"Optimal threshold ({best_threshold})")

leg = ax.legend(facecolor=SPACE, labelcolor=WHITE, fontsize=9,
                edgecolor=NEBULA)
ax.set_xlim(-0.02, 1.02)
ax.set_ylim(-0.02, 1.08)
fig.tight_layout(pad=1.5)
save_fig(fig, "02_precision_recall_curve.png")


# ── Cell 18: Chart 3 — Threshold vs Precision / Recall / F1 ─────────────────

fig, ax = plt.subplots(figsize=(9, 5), facecolor=VOID)
nasa_axes(ax, title="Threshold vs. Precision · Recall · F1  (from real y_test, y_prob)",
          xlabel="Threshold", ylabel="Score")

t_vals  = threshold_df["threshold"].values
p_vals  = threshold_df["precision_aurora"].values
r_vals  = threshold_df["recall_aurora"].values
f_vals  = threshold_df["f1_aurora"].values

ax.plot(t_vals, p_vals, "o-", color=VIOLET, lw=2, ms=5, label="Precision")
ax.plot(t_vals, r_vals, "s-", color=GREEN,  lw=2, ms=5, label="Recall")
ax.plot(t_vals, f_vals, "^-", color=TEAL,   lw=2, ms=5, label="F1-Score")

ax.axvline(best_threshold, color=ORANGE, lw=1.5, linestyle="--",
           label=f"Best F1 threshold = {best_threshold}")
ax.axvline(0.50, color=DIM, lw=1, linestyle=":",
           label="Default threshold (0.50)")

# Annotate best F1
best_f1_val = threshold_df.loc[best_f1_idx, "f1_aurora"]
best_rec_val = threshold_df.loc[best_f1_idx, "recall_aurora"]
ax.annotate(f"F1={best_f1_val:.3f}\nRecall={best_rec_val:.3f}",
            xy=(best_threshold, best_f1_val),
            xytext=(best_threshold + 0.06, best_f1_val - 0.08),
            color=TEAL, fontsize=8, fontfamily="monospace",
            arrowprops=dict(arrowstyle="->", color=TEAL, lw=1))

leg = ax.legend(facecolor=SPACE, labelcolor=WHITE, fontsize=9,
                edgecolor=NEBULA)
ax.set_xlim(0.02, 0.85)
ax.set_ylim(-0.02, 1.08)
ax.xaxis.set_minor_locator(MultipleLocator(0.025))
fig.tight_layout(pad=1.5)
save_fig(fig, "03_threshold_vs_metrics.png")


# ── Cell 19: Chart 4 — FP vs FN across thresholds ────────────────────────────

fig, ax = plt.subplots(figsize=(9, 5), facecolor=VOID)
nasa_axes(ax, title="False Positives vs. False Negatives across Thresholds",
          xlabel="Threshold", ylabel="Count")

ax.plot(t_vals, threshold_df["FP"].values, "o-", color=ORANGE, lw=2,
        ms=5, linestyle="--", label="False Positives (false alarms)")
ax.plot(t_vals, threshold_df["FN"].values, "s-", color="#ef4444", lw=2,
        ms=5, label="False Negatives (missed aurora)")

ax.axvline(best_threshold, color=TEAL, lw=1.5, linestyle="--",
           label=f"Best F1 threshold = {best_threshold}")
ax.axvline(0.50, color=DIM, lw=1, linestyle=":",
           label="Default threshold (0.50)")

# Crossover annotation
fp_arr = threshold_df["FP"].values
fn_arr = threshold_df["FN"].values
cross_idx = np.argmin(np.abs(fp_arr - fn_arr))
ax.annotate(f"FP≈FN crossover\nt≈{t_vals[cross_idx]:.2f}",
            xy=(t_vals[cross_idx], (fp_arr[cross_idx]+fn_arr[cross_idx])/2),
            xytext=(t_vals[cross_idx]+0.08, (fp_arr[cross_idx]+fn_arr[cross_idx])/2+2),
            color=YELLOW, fontsize=8, fontfamily="monospace",
            arrowprops=dict(arrowstyle="->", color=YELLOW, lw=1))

leg = ax.legend(facecolor=SPACE, labelcolor=WHITE, fontsize=9, edgecolor=NEBULA)
ax.set_xlim(0.02, 0.85)
fig.tight_layout(pad=1.5)
save_fig(fig, "04_fp_vs_fn.png")


# ── Cell 20: Chart 5 — Adjusted Confusion Matrix (best threshold) ────────────

cm_baseline = confusion_matrix(y_test, (y_prob >= 0.50).astype(int))
tn_b, fp_b, fn_b, tp_b = cm_baseline.ravel()

fig, axes = plt.subplots(1, 2, figsize=(11, 4.5), facecolor=VOID)
fig.suptitle("Confusion Matrix Comparison — Baseline vs. Optimal Threshold",
             color=WHITE, fontsize=12, fontweight="bold",
             fontfamily="monospace", y=1.02)

for ax, cm_data, title, thresh in zip(
    axes,
    [cm_baseline, cm_opt],
    ["Baseline  (threshold = 0.50)", f"Optimal  (threshold = {best_threshold})"],
    [0.50, best_threshold]
):
    ax.set_facecolor(SPACE)
    for sp in ax.spines.values(): sp.set_color(NEBULA)
    ax.set_title(title, color=WHITE, fontsize=10, fontfamily="monospace", pad=8)

    labels  = [["TN", "FP"], ["FN", "TP"]]
    colors  = [[BLUE, ORANGE], ["#ef4444", GREEN]]
    alphas  = [[0.25, 0.25], [0.35, 0.4]]

    for i in range(2):
        for j in range(2):
            val = cm_data[i, j]
            pct = val / cm_data.sum() * 100
            col = colors[i][j]
            rect = mpatches.FancyBboxPatch(
                (j+0.05, 1-i+0.05), 0.9, 0.9,
                boxstyle="round,pad=0.05",
                facecolor=col, alpha=alphas[i][j],
                edgecolor=col, linewidth=2,
                transform=ax.transData
            )
            ax.add_patch(rect)
            ax.text(j+0.5, 1-i+0.58, str(val),
                    ha="center", va="center",
                    color=col, fontsize=20, fontweight="bold",
                    fontfamily="monospace")
            ax.text(j+0.5, 1-i+0.28, f"{pct:.1f}%",
                    ha="center", va="center",
                    color=col, fontsize=9, fontfamily="monospace", alpha=0.8)
            ax.text(j+0.5, 1-i+0.12, labels[i][j],
                    ha="center", va="center",
                    color=WHITE, fontsize=9, fontfamily="monospace", alpha=0.5)

    ax.set_xlim(0, 2); ax.set_ylim(0, 2)
    ax.set_xticks([0.5, 1.5]); ax.set_yticks([0.5, 1.5])
    ax.set_xticklabels(["Predicted\nNo Aurora", "Predicted\nAurora"],
                       color=MID, fontsize=8, fontfamily="monospace")
    ax.set_yticklabels(["Actual\nAurora", "Actual\nNo Aurora"],
                       color=MID, fontsize=8, fontfamily="monospace")
    ax.tick_params(length=0)

fig.tight_layout(pad=2)
save_fig(fig, "05_confusion_matrix_comparison.png")


# ── Cell 21: Chart 6 — Feature Importance ────────────────────────────────────

importances = rf_model.feature_importances_
feat_df = pd.DataFrame({
    "feature":    features,
    "importance": importances
}).sort_values("importance", ascending=True)

# Color-code by feature group
def feat_color(name):
    if "bz" in name:    return TEAL
    if "speed" in name: return BLUE
    if "dst" in name:   return VIOLET
    if "density" in name or "proton" in name: return GREEN
    if "electric" in name or "flow_pressure" in name: return YELLOW
    return ORANGE

colors_fi = [feat_color(f) for f in feat_df["feature"]]
n = len(feat_df)
fig_h = max(6, n * 0.38)

fig, ax = plt.subplots(figsize=(9, fig_h), facecolor=VOID)
nasa_axes(ax, title="Feature Importance — Random Forest (from rf_model)",
          xlabel="Importance Score", ylabel="")
ax.set_facecolor(SPACE)

bars = ax.barh(feat_df["feature"], feat_df["importance"],
               color=colors_fi, edgecolor="none", height=0.7)

for bar, val in zip(bars, feat_df["importance"]):
    ax.text(val + 0.001, bar.get_y() + bar.get_height()/2,
            f"{val:.4f}", va="center", color=WHITE,
            fontsize=8, fontfamily="monospace")

# Legend
legend_items = [
    mpatches.Patch(color=TEAL,   label="Bz / IMF components"),
    mpatches.Patch(color=BLUE,   label="Solar wind speed"),
    mpatches.Patch(color=VIOLET, label="Dst index"),
    mpatches.Patch(color=GREEN,  label="Proton / density"),
    mpatches.Patch(color=YELLOW, label="Electric / pressure"),
    mpatches.Patch(color=ORANGE, label="Other"),
]
ax.legend(handles=legend_items, facecolor=SPACE, labelcolor=WHITE,
          fontsize=8, edgecolor=NEBULA, loc="lower right")

ax.tick_params(axis="y", colors=WHITE, labelsize=8)
ax.set_xlim(0, feat_df["importance"].max() * 1.18)
fig.tight_layout(pad=1.5)
save_fig(fig, "06_feature_importance.png")


# ── Cell 22: Chart 7 — Before / After Threshold Comparison ───────────────────

metrics_labels = ["Accuracy", "Precision", "Recall", "F1-Score"]
baseline_row   = all_results.iloc[0]
optimal_row    = all_results.iloc[1]

baseline_vals = [
    baseline_row["accuracy"],
    baseline_row["precision_aurora"],
    baseline_row["recall_aurora"],
    baseline_row["f1_aurora"],
]
optimal_vals = [
    optimal_row["accuracy"],
    optimal_row["precision_aurora"],
    optimal_row["recall_aurora"],
    optimal_row["f1_aurora"],
]

x     = np.arange(len(metrics_labels))
width = 0.35

fig, ax = plt.subplots(figsize=(9, 5.5), facecolor=VOID)
nasa_axes(ax, title="Before vs. After Threshold Tuning",
          xlabel="Metric", ylabel="Score")

bars1 = ax.bar(x - width/2, baseline_vals, width,
               label=f"Baseline (t=0.50)", color=BLUE, alpha=0.7,
               edgecolor=SPACE, linewidth=0.5)
bars2 = ax.bar(x + width/2, optimal_vals,  width,
               label=f"Tuned (t={best_threshold})", color=GREEN, alpha=0.8,
               edgecolor=SPACE, linewidth=0.5)

for bar in bars1:
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.008,
            f"{bar.get_height():.3f}", ha="center", va="bottom",
            color=BLUE, fontsize=8, fontfamily="monospace")
for bar in bars2:
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.008,
            f"{bar.get_height():.3f}", ha="center", va="bottom",
            color=GREEN, fontsize=8, fontfamily="monospace")

# Delta annotations
for i, (b, o) in enumerate(zip(baseline_vals, optimal_vals)):
    delta = o - b
    sign  = "+" if delta >= 0 else ""
    col   = GREEN if delta > 0 else ORANGE
    ax.text(i, max(b, o) + 0.055, f"{sign}{delta:.3f}",
            ha="center", va="bottom", color=col,
            fontsize=8, fontfamily="monospace", fontweight="bold")

ax.set_xticks(x)
ax.set_xticklabels(metrics_labels, color=WHITE, fontsize=10,
                   fontfamily="monospace")
ax.set_ylim(0, 1.18)
ax.axhline(1.0, color=DIM, lw=0.5, linestyle=":")
leg = ax.legend(facecolor=SPACE, labelcolor=WHITE, fontsize=9,
                edgecolor=NEBULA)
fig.tight_layout(pad=1.5)
save_fig(fig, "07_before_after_comparison.png")


# ── Cell 23: Chart 8 — Full Dashboard (all charts in one figure) ─────────────

fig = plt.figure(figsize=(18, 14), facecolor=VOID)
fig.suptitle(
    "AuroraAI  ·  Threshold Tuning Mission Control  ·  NASA OMNIWeb Space Weather ML",
    color=WHITE, fontsize=14, fontweight="bold",
    fontfamily="monospace", y=0.995
)

gs = gridspec.GridSpec(3, 3, figure=fig, hspace=0.42, wspace=0.32,
                       left=0.06, right=0.97, top=0.96, bottom=0.05)

# 1 — ROC
ax1 = fig.add_subplot(gs[0, 0])
nasa_axes(ax1, "ROC Curve", "FPR", "TPR")
ax1.fill_between(fpr_vals, tpr_vals, alpha=0.12, color=GREEN)
ax1.plot(fpr_vals, tpr_vals, color=GREEN, lw=2,
         label=f"AUC={roc_auc_val:.4f}")
ax1.plot([0,1],[0,1],"--",color=DIM,lw=1)
ax1.scatter(fpr_vals[opt_idx], tpr_vals[opt_idx], color=ORANGE, s=60, zorder=5)
ax1.legend(facecolor=SPACE,labelcolor=WHITE,fontsize=8,edgecolor=NEBULA)
ax1.set_xlim(-0.02,1.02); ax1.set_ylim(-0.02,1.08)

# 2 — PR curve
ax2 = fig.add_subplot(gs[0, 1])
nasa_axes(ax2, "Precision-Recall Curve", "Recall", "Precision")
ax2.fill_between(rec_vals, prec_vals, alpha=0.12, color=TEAL)
ax2.plot(rec_vals, prec_vals, color=TEAL, lw=2,
         label=f"PR-AUC={pr_auc_val:.4f}")
ax2.axhline(no_skill, color=DIM, lw=1, linestyle="--")
ax2.scatter(rec_vals[opt_pr_idx], prec_vals[opt_pr_idx],
            color=ORANGE, s=60, zorder=5)
ax2.legend(facecolor=SPACE,labelcolor=WHITE,fontsize=8,edgecolor=NEBULA)
ax2.set_xlim(-0.02,1.02); ax2.set_ylim(-0.02,1.08)

# 3 — Threshold vs metrics
ax3 = fig.add_subplot(gs[0, 2])
nasa_axes(ax3, "Threshold vs Metrics", "Threshold", "Score")
ax3.plot(t_vals, p_vals, "o-", color=VIOLET, lw=1.5, ms=4, label="Precision")
ax3.plot(t_vals, r_vals, "s-", color=GREEN,  lw=1.5, ms=4, label="Recall")
ax3.plot(t_vals, f_vals, "^-", color=TEAL,   lw=1.5, ms=4, label="F1")
ax3.axvline(best_threshold, color=ORANGE, lw=1.2, linestyle="--")
ax3.axvline(0.50, color=DIM, lw=0.8, linestyle=":")
ax3.legend(facecolor=SPACE,labelcolor=WHITE,fontsize=7,edgecolor=NEBULA)
ax3.set_xlim(0.02,0.85); ax3.set_ylim(-0.02,1.08)

# 4 — FP vs FN
ax4 = fig.add_subplot(gs[1, 0])
nasa_axes(ax4, "FP vs FN across Thresholds", "Threshold", "Count")
ax4.plot(t_vals, threshold_df["FP"].values, "o--", color=ORANGE, lw=1.5, ms=4, label="False Positives")
ax4.plot(t_vals, threshold_df["FN"].values, "s-",  color="#ef4444", lw=1.5, ms=4, label="False Negatives")
ax4.axvline(best_threshold, color=TEAL, lw=1.2, linestyle="--")
ax4.legend(facecolor=SPACE,labelcolor=WHITE,fontsize=8,edgecolor=NEBULA)

# 5 — F1 bar
ax5 = fig.add_subplot(gs[1, 1])
nasa_axes(ax5, "F1 Score per Threshold", "Threshold", "F1")
bar_colors = [GREEN if abs(v-f_vals.max())<0.001 else BLUE+"88" for v in f_vals]
ax5.bar(t_vals, f_vals, width=0.04, color=bar_colors, edgecolor="none")
ax5.axvline(best_threshold, color=ORANGE, lw=1.2, linestyle="--",
            label=f"Best t={best_threshold}")
ax5.legend(facecolor=SPACE,labelcolor=WHITE,fontsize=8,edgecolor=NEBULA)
ax5.set_ylim(0, 1.1)

# 6 — Before/After
ax6 = fig.add_subplot(gs[1, 2])
nasa_axes(ax6, "Before vs. After Tuning", "Metric", "Score")
xb = np.arange(len(metrics_labels))
ax6.bar(xb-0.18, baseline_vals, 0.34, color=BLUE, alpha=0.75, label="Baseline (0.50)")
ax6.bar(xb+0.18, optimal_vals,  0.34, color=GREEN, alpha=0.8,  label=f"Tuned ({best_threshold})")
ax6.set_xticks(xb)
ax6.set_xticklabels(["Acc","Prec","Rec","F1"], color=WHITE, fontsize=9)
ax6.set_ylim(0,1.18)
ax6.legend(facecolor=SPACE,labelcolor=WHITE,fontsize=8,edgecolor=NEBULA)

# 7 — Feature importance (top 12)
ax7 = fig.add_subplot(gs[2, :])
nasa_axes(ax7, "Feature Importance — Random Forest", "Importance Score", "")
top_feat = feat_df.tail(12)
top_colors = [feat_color(f) for f in top_feat["feature"]]
ax7.barh(top_feat["feature"], top_feat["importance"],
         color=top_colors, edgecolor="none", height=0.7)
for i, (name, val) in enumerate(zip(top_feat["feature"], top_feat["importance"])):
    ax7.text(val+0.0005, i, f"{val:.4f}", va="center",
             color=WHITE, fontsize=7.5, fontfamily="monospace")
ax7.tick_params(axis="y", colors=WHITE, labelsize=8)
ax7.set_xlim(0, top_feat["importance"].max()*1.16)

save_fig(fig, "08_full_dashboard.png")


# ── Cell 24: Print final summary ─────────────────────────────────────────────

print()
print("=" * 62)
print("  AURORAAI — THRESHOLD TUNING COMPLETE")
print("=" * 62)
print(f"  Best threshold        : {best_threshold}")
print(f"  Recall (baseline)     : {baseline_row['recall_aurora']:.4f}")
print(f"  Recall (tuned)        : {optimal_row['recall_aurora']:.4f}  ← improvement")
print(f"  Precision (baseline)  : {baseline_row['precision_aurora']:.4f}")
print(f"  Precision (tuned)     : {optimal_row['precision_aurora']:.4f}")
print(f"  F1 (baseline)         : {baseline_row['f1_aurora']:.4f}")
print(f"  F1 (tuned)            : {optimal_row['f1_aurora']:.4f}  ← improvement")
print(f"  ROC-AUC (unchanged)   : {optimal_row['roc_auc']:.4f}")
print()
print("  Saved charts:")
for fname in ["01_roc_curve.png","02_precision_recall_curve.png",
              "03_threshold_vs_metrics.png","04_fp_vs_fn.png",
              "05_confusion_matrix_comparison.png","06_feature_importance.png",
              "07_before_after_comparison.png","08_full_dashboard.png"]:
    print(f"    → {os.path.join(OUTPUT_DIR, fname)}")
print()
print(f"  threshold_df    : {len(threshold_df)} rows × {len(threshold_df.columns)} cols")
print(f"  best_threshold  : {best_threshold}")
print(f"  all_results     : {len(all_results)} models compared")
print("=" * 62)
```

```python
"""
AuroraAI — Decision Boundary: Aurora Probability Field
Bz_GSM vs Solar Wind Speed (Feature Space Projection)

Reproduces the probability contour map from the AuroraAI dashboard.
Run in Anaconda (Jupyter Notebook or JupyterLab):
    conda activate base
    pip install matplotlib numpy  # if not already installed
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.lines import Line2D

# ── Reproducible random state ──────────────────────────────────────────────
rng = np.random.default_rng(42)

# ══════════════════════════════════════════════════════════════════════════
# 1.  SYNTHETIC FEATURE-SPACE DATA
#     X-axis : Bz_GSM (nT)   — negative = southward = aurora-favorable
#     Y-axis : Solar Wind Speed (km/s)
#     Aurora class: ~9.5% of data (mirrors real class imbalance)
# ══════════════════════════════════════════════════════════════════════════
N_total  = 340
N_aurora = int(N_total * 0.095)          # ~32 aurora points (red)
N_noaur  = N_total - N_aurora            # ~308 non-aurora (purple)

# Non-aurora cluster: centred around Bz≈0, speed≈450 km/s
bz_no    = rng.normal(loc=0,    scale=6,   size=N_noaur)
spd_no   = rng.normal(loc=450,  scale=55,  size=N_noaur)

# Aurora cluster: negative Bz, elevated speed — upper-left scatter
bz_aur   = rng.uniform(low=-25, high=5,   size=N_aurora)
spd_aur  = rng.normal(loc=600,  scale=90, size=N_aurora)

# ══════════════════════════════════════════════════════════════════════════
# 2.  PROBABILITY CONTOUR FIELD
#     P(aurora) increases with: lower Bz, higher speed
#     Simple logistic surface for visualisation only — not real model output
# ══════════════════════════════════════════════════════════════════════════
bz_grid  = np.linspace(-30, 20,  300)
spd_grid = np.linspace(280, 820, 300)
BZ, SPD  = np.meshgrid(bz_grid, spd_grid)

# Aurora probability proxy surface
Z = 1 / (1 + np.exp(0.25 * BZ - 0.008 * (SPD - 400)))
Z = np.clip(Z, 0, 1)

# ══════════════════════════════════════════════════════════════════════════
# 3.  CUSTOM COLOUR MAPS
# ══════════════════════════════════════════════════════════════════════════
# Contour fill: blue (low prob) → teal → green → red (high prob)
contour_cmap = LinearSegmentedColormap.from_list(
    "aurora_prob",
    ["#2563eb", "#0891b2", "#059669", "#65a30d", "#ca8a04", "#dc2626"],
    N=256
)

# ══════════════════════════════════════════════════════════════════════════
# 4.  PLOT  (enlarged: 14 × 8 inches)
# ══════════════════════════════════════════════════════════════════════════
fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor("#f8f9fc")
ax.set_facecolor("#f8faff")

# — Filled probability contours ——————————————————————————————————————
cf = ax.contourf(BZ, SPD, Z, levels=20, cmap=contour_cmap, alpha=0.22)

# — Contour lines ————————————————————————————————————————————————————
cl = ax.contour(
    BZ, SPD, Z,
    levels=[0.1, 0.3, 0.5, 0.7, 0.85],
    colors=["#2563eb", "#0891b2", "#059669", "#ca8a04", "#dc2626"],
    linewidths=1.2,
    alpha=0.55
)
ax.clabel(cl, fmt=lambda v: f"{v:.0%}", fontsize=8, colors="#334155")

# — Scatter: non-aurora (purple) ————————————————————————————————————
ax.scatter(
    bz_no, spd_no,
    s=38, c="#7c3aed", alpha=0.72,
    edgecolors="#ede9fe", linewidths=0.6,
    zorder=4, label="No Aurora (Kp < 5)"
)

# — Scatter: aurora events (red) ————————————————————————————————————
ax.scatter(
    bz_aur, spd_aur,
    s=52, c="#dc2626", alpha=0.84,
    edgecolors="#fee2e2", linewidths=0.8,
    zorder=5, label="Aurora (Kp ≥ 5)"
)

# — Vertical threshold line at Bz = 0 ——————————————————————————————
ax.axvline(0, color="#94a3b8", linewidth=1, linestyle="--", alpha=0.5, zorder=3)

# — Annotation labels ——————————————————————————————————————————————
ax.text(-26, 760, "High Kp — aurora",
        fontsize=10, fontweight="bold", color="#7c2d12",
        bbox=dict(boxstyle="round,pad=0.3", facecolor="#fef3c7",
                  edgecolor="#fcd34d", alpha=0.85))

ax.text(2, 420, "Low Kp — no aurora",
        fontsize=10, fontweight="bold", color="#1e293b",
        bbox=dict(boxstyle="round,pad=0.3", facecolor="white",
                  edgecolor="#e2e8f0", alpha=0.85))

# — Colour bar (probability scale) ————————————————————————————————
cbar = fig.colorbar(cf, ax=ax, pad=0.02, shrink=0.75)
cbar.set_label("Aurora Probability", fontsize=10, color="#334155")
cbar.ax.yaxis.set_tick_params(color="#94a3b8")
plt.setp(cbar.ax.yaxis.get_ticklabels(), color="#64748b", fontsize=9)

# ══════════════════════════════════════════════════════════════════════════
# 5.  AXES STYLING
# ══════════════════════════════════════════════════════════════════════════
ax.set_xlabel("Bz_GSM (nT)  ← southward favors aurora", fontsize=11, color="#334155", labelpad=8)
ax.set_ylabel("Solar Wind Speed (km/s)", fontsize=11, color="#334155", labelpad=8)
ax.set_title(
    "AuroraAI · Decision Boundary — Aurora Probability Field\n"
    "Bz_GSM vs Solar Wind Speed  (Feature Space Projection)",
    fontsize=13, fontweight="bold", color="#0f172a", pad=14
)

ax.set_xlim(-30, 20)
ax.set_ylim(280, 820)
ax.tick_params(colors="#94a3b8", labelsize=9)
for spine in ax.spines.values():
    spine.set_edgecolor("#e2e8f0")
ax.grid(color="#e2e8f0", linewidth=0.6, alpha=0.8)

# ══════════════════════════════════════════════════════════════════════════
# 6.  LEGEND
# ══════════════════════════════════════════════════════════════════════════
legend_handles = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#dc2626',
           markeredgecolor='#fee2e2', markersize=10, label='Aurora (Kp ≥ 5)'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#7c3aed',
           markeredgecolor='#ede9fe', markersize=10, label='No Aurora (Kp < 5)'),
    mpatches.Patch(facecolor='#2563eb', alpha=0.4, label='Low probability region'),
    mpatches.Patch(facecolor='#dc2626', alpha=0.4, label='High probability region'),
]
ax.legend(
    handles=legend_handles,
    loc='lower left', fontsize=9,
    framealpha=0.9, edgecolor="#e2e8f0",
    facecolor="white"
)

# ══════════════════════════════════════════════════════════════════════════
# 7.  CAPTION
# ══════════════════════════════════════════════════════════════════════════
fig.text(
    0.5, 0.01,
    "Probability surface is a logistic approximation for visualisation. "
    "Actual model: Random Forest · ROC-AUC = 0.9769 · Tuned threshold = 0.40",
    ha='center', fontsize=8, color="#94a3b8"
)

plt.tight_layout(rect=[0, 0.03, 1, 1])

# ── Save ──────────────────────────────────────────────────────────────────
plt.savefig("aurora_decision_boundary.png", dpi=150, bbox_inches="tight",
            facecolor=fig.get_facecolor())
print("Saved: aurora_decision_boundary.png")

plt.show()
```
